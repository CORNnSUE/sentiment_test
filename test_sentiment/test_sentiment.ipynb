```python
import pandas as pd
json_file_path = "all.json"
df = pd.read_json(json_file_path)
df
```

```python
df['label'] = df['label'].map(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)
df.dropna(subset=['label'], inplace=True)
print(df.head())
```

```python
json_file_path = "duoc.json"
df_large = pd.read_json(json_file_path)
```

```python
new_df = df[["text", "label"]]
new_df.head()
```

```python
new_df_large = df_large[["title", "category"]]
print(new_df_large.head())
```

```python
new_df_large.dropna(subset=['category'], inplace=True)
new_df_large.tail()
```

```python
new_df_large.rename(columns={'title': 'text', 'category': 'label'}, inplace=True)
new_df_large.tail()
```

```python
data_raw = pd.concat([new_df, new_df_large], ignore_index=True)
label_counts = data_raw['label'].value_counts()
label_counts
```

```python
max_samples_per_label = 1500

data = pd.DataFrame(columns=['text', 'label'])

for label in data_raw['label'].unique():
    label_data = data_raw[data_raw['label'] == label]
    
    if len(label_data) > max_samples_per_label:
        label_data = label_data.sample(max_samples_per_label, random_state=42)
    
    data = pd.concat([data, label_data], ignore_index=True)
data
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

data['label_encoded'] = label_encoder.fit_transform(data['label'])
data
```

```python
unique_encoded_values = data['label_encoded'].unique()

for value in unique_encoded_values:
    label = data[data['label_encoded'] == value]['label'].iloc[0]
    print(f"{value} -> {label}")
```

# XGboost

```python
from xgboost import XGBClassifier
xgb_classifier = XGBClassifier(n_estimators=100, random_state=42)
```

```python
for train_index, test_index in stratified_kfold.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    xgb_classifier.fit(X_train, y_train)
    
    y_pred = xgb_classifier.predict(X_test)


    wrong_indices = test_index[y_test != y_pred]
    
    wrong_data = data.iloc[wrong_indices]
    wrong_data.loc[y_test != y_pred, 'predicted_label'] = y_pred[y_test != y_pred]    
    wrong_df = pd.concat([wrong_df, wrong_data], ignore_index=True)
    
    classification_rep = classification_report(y_test, y_pred, output_dict=True)
    f1_scores.append(classification_rep['macro avg']['f1-score'])
    precision_scores.append(classification_rep['macro avg']['precision'])
    recall_scores.append(classification_rep['macro avg']['recall'])
    accuracy_scores.append(classification_rep['accuracy'])
```

```python
print("Average F1 Score:", sum(f1_scores) / len(f1_scores))
print("Average Precision:", sum(precision_scores) / len(precision_scores))
print("Average Recall:", sum(recall_scores) / len(recall_scores))
print("Average Accuracy:", sum(accuracy_scores) / len(accuracy_scores))
```

# Neural Network

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import classification_report
```

```python
tfidf_vectorizer = TfidfVectorizer()
X = tfidf_vectorizer.fit_transform(data['text'])
y = data['label_encoded']
unique_labels = data['label_encoded'].unique()
```

```python
f1_scores = {label: [] for label in y.unique()}
precision_scores = {label: [] for label in y.unique()}
recall_scores = {label: [] for label in y.unique()}
```

```python
X = torch.tensor(X.toarray(), dtype=torch.float32)
y = torch.tensor(y.values, dtype=torch.int64)
```

```python
class SimpleNN(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out
```

```python
input_size = X.shape[1]
hidden_size = 64 
num_classes = len(data['label'].unique()) 

model = SimpleNN(input_size, hidden_size, num_classes)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# cross-validation
for train_index, test_index in stratified_kfold.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    train_dataset = TensorDataset(X_train, y_train)
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    
    # train
    model.train()
    for epoch in range(10):
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

    model.eval()
    with torch.no_grad():
        outputs = model(X_test)
        _, predicted = torch.max(outputs, 1)
        
        classification_rep = classification_report(y_test, predicted, target_names=unique_labels, output_dict=True)
        
        for label in unique_labels:
            precision = classification_rep[label]['precision']
            recall = classification_rep[label]['recall']
            f1 = classification_rep[label]['f1-score']
            
            precision_scores[label].append(precision)
            recall_scores[label].append(recall)
            f1_scores[label].append(f1)
```

```python
torch.save(model.state_dict(), "model_NN.pth")
```

```python
import pandas as pd
import numpy as np

results = pd.DataFrame(columns=['NEG', 'NEU', 'POS'], index=['Precision', 'Recall', 'F1-Score'])

label_mapping = {0: 'NEG', 1: 'NEU', 2: 'POS'}

for label in unique_labels:
    avg_precision = np.mean(precision_scores[label])
    avg_recall = np.mean(recall_scores[label])
    avg_f1 = np.mean(f1_scores[label])

    mapped_label = label_mapping[label]
    results.at['Precision', mapped_label] = avg_precision
    results.at['Recall', mapped_label] = avg_recall
    results.at['F1-Score', mapped_label] = avg_f1

print(results)
```
